```python
import pandas as pd
from bs4 import BeautifulSoup

# Supongamos que ya tienes el contenido HTML cargado en `soup`
amazon_data = pd.DataFrame(columns=["Date", "Open", "High", "Low", "Close", "Adj Close", "Volume"])

for row in soup.find("tbody").find_all("tr"):
    col = row.find_all("td")

    # Extraer cada campo desde las columnas (asumiendo el orden correcto en la tabla)
    date = col[0].text.strip()
    Open = col[1].text.strip()
    high = col[2].text.strip()
    low = col[3].text.strip()
    close = col[4].text.strip()
    adj_close = col[5].text.strip()
    volume = col[6].text.strip()

    # Agregar la fila al DataFrame
    amazon_data = pd.concat([
        amazon_data,
        pd.DataFrame({
            "Date": [date],
            "Open": [Open],
            "High": [high],
            "Low": [low],
            "Close": [close],
            "Adj Close": [adj_close],
            "Volume": [volume]
        })
    ], ignore_index=True)
```
